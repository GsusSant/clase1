# Ejercicio 3

Dada una lista y un número, cree una función que devuelva dos parámetros:

1) Si está dicho número en esa lista. En dicho caso Devolverá `TRUE`. En caso contrario, `False`.

2) Si la longitud de la lista coincide con dicho número. En dicho caso devolverá `TRUE`. En caso contrario, `False`.

````python
def doble_salida(lista_numeros, numero_comprobar):
    esta_num_presente = True if numero_comprobar in lista_numeros else False
    longitud_igual = True if len(lista_numeros) == numero_comprobar else False
    return esta_num_presente, longitud_igual

print(doble_salida([1, 3, 5, 8, 10, 7], 5))
````

# Ejercicio 7
Cree una función para contar la cantidad de números negativos de una lista.


````python
lista = [1, -3.3, -4, 5, 2, 18, -9, 11, -5.5]

#THE FUNCTION IS DEFINED
def contador_neg(lista): 
    #THE VARIABLE IS INITIALISED TO 0
    contador_negativos = 0
    
    #ITERATES OVER EACH NUMBER IN THE LIST AND INCREMENTS THE VALUE EACH TIME A NEGATIVE NUMBER IS ENCOUNTERED.
    for numero in lista: 
        if numero < 0 : 
            contador_negativos += 1
        
    return contador_negativos


resultado = contador_neg(lista)
print(f"La cantidad de números negativos en la lista es de: {resultado}")
````

````python
def contar_negativos(lista_numeros): 
    if not lista_numeros: return[]
    return [len([numero for numero in lista_numeros if numero < 0])]
lista = [1, -3.3, -4, 5, 2, 18, -9, 11, -5.5]

print(contar_negativos(lista))
````

# Ejercicio 8

Cree una función para contar la cantidad de números impares de una lista.


````python
lista = [1, 3, 7, -3, -3, -4]

#THE FUNCTION IS DEFINED
def contador_impares(lista): 
    contador_impares = 0
    
    for numero in lista: 
        if (numero%2 != 0): 
            contador_impares += 1
    return contador_impares


resultado = contador_impares(lista)
print(f"La cantidad de números impares en la lista es de: {resultado}")
````

# Ejercicio 9

Dada una palabra, cuente el número de vocales que tiene.



````python
# CADENA DE TEXTO
frase = "hola BuéNNOosS DíAsh"

# CONTADOR DE VOCALES
contador_vocales = 0

# DEFINIMOS VOCALES
vocales = "aeiouAEIOUáéíóúÁÉÍÓÚ"

# ITERAR SOBRE CADA CARACTER EN LA CADENA
for caracter in frase:
  #VERIFICAR SI CARACTER ES VOCAL

  if caracter in vocales:
    contador_vocales += 1

# MOSTRAR RESULTADO
print(f"la cadena ingresada contiene {contador_vocales} vocales")
````

````python

def vocales_contar(palabra): 
    vocales = ['a', 'e', 'i', 'o', 'u', 'á', 'é', 'í', 'ó', 'ú', 'A', 'E', 'I', 'O', 'U','Á', 'É', 'Í', 'Ò', 'Ù']
    total_vocales = 0
    
    for vocal in vocales: 
        if vocal in palabra:
            total_vocales += palabra.count(vocal)
            
    return total_vocales

print(vocales_contar("DíAsh"))
    
````

# Ejercicio de ampliación 1
Modifique el programa de la contraseña para dar al usuario solo 3 intentos. Si pasados dichos intentos el usuario no ha escrito bien la contraseña el programa acabará.

````python
contrasena_correcta = "upgrade"
 
for i in range(3):
    contrasena_ingresada = input("Por favor, ingresa tu contraseña: ")
   
    if contrasena_ingresada == contrasena_correcta:
        print("Bienvenido al sistema!")
        break  
    else:
        print("Contraseña incorrecta.")
        if i < 2:
            print("Inténtalo de nuevo.")
        else:
            print("Has agotado tus 3 intentos. Acceso denegado.")
````

````python
max_attempts = 3

for attempt in range(1, max_attempts + 1):
    password = input("Por favor, ingrese la contraseña: ")

    if password == "secret":
        print("Gracias. Ha ingresado la contraseña correcta.")
        break
    else:
        print(f"Lo sentimos, la contraseña es incorrecta. Intentos restantes: {max_attempts - attempt}")

    if attempt != max_attempts:
        print("Número máximo de intentos alcanzado. Cerrando el programa.")
````
